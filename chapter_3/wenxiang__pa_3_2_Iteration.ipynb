1. Write a function that takes in information about a penguin, and returns one of the following definitions:

*   "Big Mouth Billy": Male penguins with bill length times bill depth greater than 800.
*   "Dainty Daisy": Female penguins with flipper length less than 5% of body mass.
*   "Average Adelie:" Any Adelie penguins that do not fall into either category.
*   "Other": Any penguins that do not fall into any of the categories.



```python
penguins = p.load_penguins()
penguins.head()
```

- Write the Big Mouth Billy function

```python
def penguin_category(species, bill_length_mm, bill_depth_mm, flipper_length_mm, body_mass_g, sex):
  """
  Categorize each observation of penguins and return of the category

  Arguments
  ---------
  species, bill_length_mm, bill_depth_mm, flipper_length_mm, body_mass_g, sex

  Return:
  "Big Mouth Billy": Male penguins with bill length times bill depth greater than 800.
  "Dainty Daisy": Female penguins with flipper length less than 5% of body mass.
  "Average Adelie:" Any Adelie penguins that do not fall into either category.
  "Other": Any penguins that do not fall into any of the categories.
  """
  if sex == 'male' and bill_length_mm * bill_depth_mm > 800:
    return "Big Mouth Billy"
  if sex == 'female' and flipper_length_mm < body_mass_g * 0.05:
    return "Dainty Daisy"
  if species == 'Adelie' and not ((sex == 'male' and bill_length_mm * bill_depth_mm > 800) or (sex == 'female' and flipper_length_mm < body_mass_g * 0.05)):
    return "Average Adelie"
  else:
    return "Other"
```

```python
print(penguin_category('Gebtoo', 40, 21, 195.0, 3250, 'male'))
```

```python
print(penguin_category('Gentoo', 40, 20, 0.1, 3750, 'female'))
```

2. Use an iterable function to create a new variable called `category_name` that adds these labels.

- Map() method

```python
category_name = list(map(penguin_category,
                    penguins['species'],
                    penguins['bill_length_mm'],
                    penguins['bill_depth_mm'],
                    penguins['flipper_length_mm'],
                    penguins['body_mass_g'],
                    penguins['sex']
                    )
)
print("".join(category_name))
```

```python
penguins = penguins.assign(category_name = category_name)
```

```python
penguins.head()
```

```python
penguins['category'] = penguins.apply(lambda x: penguin_category(
    x['species'],
    x['bill_length_mm'],
    x['bill_depth_mm'],
    x['flipper_length_mm'],
    x['body_mass_g'],
    x['sex']
), axis = 1
)
```
