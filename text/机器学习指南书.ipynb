# 🧠 一句话比喻：常见机器学习模型速查表

| 模型 | 一句话性格比喻 | 主场优势 |
|------|-----------------|----------|
| **Logistic Regression** | 📐「线性思维严谨的律师」 | 解释性强，线性边界清晰 |
| **Naive Bayes**         | 📦「条件独立的快递分拣员」 | 高维稀疏文本 & one-hot 数据之神 |
| **SVM**                 | ✂️「边界控的裁缝师」 | 中小数据，边界分明，维度高也不怕 |
| **Random Forest**       | 🌲「多数决定一切的森林投票系统」 | 稳定、鲁棒、不会过拟合得太离谱 |
| **AdaBoost**            | 🎯「专治小错的精准打击手」 | 迭代聚焦难点，适合噪声少的数据 |
| **XGBoost**             | 🏆「打 tabular 比赛的冠军选手」 | 表格类数据王者，强大又快 |
| **LightGBM**            | 🚀「XGBoost 的极速版」 | 超大规模数据，训练飞快 |
| **MLP（浅神经网）**     | 🔌「会拉非线性电路的工程师」 | 表格数据中的非线性建模专家 |
| **CNN**                 | 👁️「看图识物的视觉大师」 | 图像/2D结构识别无敌 |
| **RNN / LSTM**          | ⏳「有记忆的时序诗人」 | 时间序列、语言序列首选 |
| **Transformer / BERT** | 🧠「懂上下文语义的语言学家」 | 文本理解和生成的现役王者 |
| **KNN**                 | 🧍「离你最近的那个邻居」 | 简单直观，小数据好用但不快 |
| **Stacking**            | 🧩「模型界的融合 DJ」 | 汇聚多模多才模型，取长补短 |

#### 1. 代码

```python
import ipywidgets as widgets
from IPython.display import display, Markdown

# 定义选项
feature_type = widgets.Dropdown(
    options=['数值型为主', '类别型为主（one-hot）', '混合型', '文本', '图像', '时间序列'],
    description='特征类型:',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='400px')
)

data_size = widgets.Dropdown(
    options=['少量数据（<10K）', '中等数据（10K-100K）', '大量数据（>100K）'],
    description='样本规模:',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='400px')
)

task_type = widgets.Dropdown(
    options=['二分类', '多分类', '回归'],
    description='任务类型:',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='400px')
)

# 输出函数
def recommend_model(feature, size, task):
    recs = []
    if task == '二分类':
        if feature == '类别型为主（one-hot）':
            recs = ['✅ BernoulliNB', '✅ XGBoost', '✅ Logistic Regression']
        elif feature == '数值型为主':
            recs = ['✅ Logistic Regression', '✅ Random Forest', '✅ SVM']
        elif feature == '混合型':
            recs = ['✅ XGBoost', '✅ LightGBM', '✅ MLP']
        elif feature == '文本':
            recs = ['✅ Naive Bayes', '✅ SVM', '✅ Transformer（BERT）']
        elif feature == '图像':
            recs = ['✅ CNN', '✅ Transfer Learning']
        elif feature == '时间序列':
            recs = ['✅ RNN / LSTM', '✅ Temporal CNN']
    elif task == '回归':
        recs = ['✅ XGBoost Regressor', '✅ Random Forest Regressor', '✅ Ridge / Lasso']
    elif task == '多分类':
        if feature == '图像':
            recs = ['✅ CNN + Softmax', '✅ EfficientNet']
        else:
            recs = ['✅ XGBoost', '✅ MLP', '✅ Random Forest']

    display(Markdown(f"### 🔍 推荐模型：\n- " + '\n- '.join(recs)))

# 交互控件
tool_ui = widgets.VBox([feature_type, data_size, task_type])
ui_out = widgets.interactive_output(recommend_model, {
    'feature': feature_type,
    'size': data_size,
    'task': task_type
})


```

#### 2.选择面板

```python
display(Markdown("## 🤖 模型推荐器"))
display(tool_ui, ui_out)
```

#### 3. 一键分类器 Pipeline A

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay

from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier

from sklearn.datasets import make_classification
X, y = make_classification(n_samples=5000, n_features=20, n_informative=15,
                           n_redundant=5, n_classes=2, random_state=42)

scaler = StandardScaler()
X = scaler.fit_transform(X)

models = {
    'BernoulliNB': BernoulliNB(),
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'AdaBoost': AdaBoostClassifier(n_estimators=100, random_state=42),
    'MLP': MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=300, random_state=42),
    'XGBoost': XGBClassifier(eval_metric='logloss', use_label_encoder=False, random_state=42)
}

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = []
conf_matrices = {}

for name, model in models.items():
    print(f"Running cross-validation for {name}...")
    acc_list, f1_list, auc_list, prec_list, rec_list = [], [], [], [], []
    conf_matrix_sum = np.zeros((2, 2), dtype=int)
    start_time = time.time()

    for train_idx, test_idx in kf.split(X, y):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else np.zeros_like(y_pred)

        acc_list.append(accuracy_score(y_test, y_pred))
        f1_list.append(f1_score(y_test, y_pred))
        auc_list.append(roc_auc_score(y_test, y_proba))
        prec_list.append(precision_score(y_test, y_pred))
        rec_list.append(recall_score(y_test, y_pred))
        conf_matrix_sum += confusion_matrix(y_test, y_pred)

    elapsed = time.time() - start_time

    results.append({
        'Model': name,
        'Accuracy (mean)': np.mean(acc_list),
        'F1 Score (mean)': np.mean(f1_list),
        'Precision (mean)': np.mean(prec_list),
        'Recall (mean)': np.mean(rec_list),
        'AUC (mean)': np.mean(auc_list),
        'CV Time (s)': elapsed
    })
    conf_matrices[name] = conf_matrix_sum

results_df = pd.DataFrame(results).sort_values(by='F1 Score (mean)', ascending=False)
results_df.reset_index(drop=True, inplace=True)

print("CV Result：")
print(results_df)

results_df.plot(x='Model', y=['Accuracy (mean)', 'F1 Score (mean)', 'Precision (mean)', 'Recall (mean)', 'AUC (mean)'], kind='bar', figsize=(14, 6))
plt.title("CV Performance Comparison")
plt.ylabel("Score")
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# CM
for name, cm in conf_matrices.items():
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
    disp.plot(cmap='Blues')
    plt.title(f"CM - {name}")
    plt.tight_layout()
    plt.show()

# 保存结果为 CSV
#results_df.to_csv("model_cv_comparison_results.csv", index=False)
#print("交叉验证结果已保存为 model_cv_comparison_results.csv")
```
